```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import image 
from matplotlib import pyplot as plt 
```

```python
#import the data
#index col = [0] makes sures another index column is not added

game_data = pd.read_csv(r'C:\Users\akobe\OneDrive\Desktop\Lighthouse\After\NHL-Stats\Data\game_data.csv', index_col = [0]) 
```

```python
#colour palette as dictionary 
#Hex colour codes for team, 0(away) = MTL, 1(home) = TOR
team_palette = {0 : '#AF1E2D',
           1 : '#00205B'}
```

```python
#plotting coordinates on scatterplot 
sns.scatterplot(data = game_data, x = 'x_coor', y = 'y_coor', hue = 'home_team', palette = team_palette)
plt.legend(title='Teams', loc='upper left', bbox_to_anchor=(1, 1))
```

```python
#colour palette as dictionary 
#Hex colour codes for team, 0(away) = MTL, 1(home) = TOR
event_palette = {'SHOT': '#3366FF',
            'GOAL': '#009933',
            'MISSED_SHOT': '#FF0000',
            'BLOCKED_SHOT': '#FF944D'}
```

```python
#plotting coordinates on scatterplot, by event_type
sns.scatterplot(data = game_data, x = 'x_coor', y = 'y_coor', hue = 'event_type', palette = event_palette)
plt.legend(title='Shot Outcome', loc='upper left', bbox_to_anchor=(1, 1))
```

## Plot on Rink

```python
game_data.head(2)
```

```python
#set the inverse multiplier
inverse = -1
```

```python
#prepping the data - need all shots to be going in the same direction - right side 
#x_coordinate
game_data['x_coor_2'] = np.where(game_data['rinkside'] == 'left',              #condition 
                                                game_data['x_coor'] * inverse, #what to assign when condition is true
                                                game_data['x_coor']            #what to assign when the conditon is false 
                                            )
```

```python
#prepping the data - need all shots to be going in the same direction - right side 
#y_coordinate
game_data['y_coor_2'] = np.where(game_data['rinkside'] == 'left',              #condition 
                                                game_data['y_coor'] * inverse, #what to assign when condition is true
                                                game_data['y_coor']            #what to assign when the conditon is false 
                                            )
```

```python
#plotting coordinates on scatterplot 
sns.scatterplot(data = game_data, x = 'x_coor_2', y = 'y_coor_2', hue = 'home_team', palette = team_palette)
plt.legend(title='Teams', loc='upper left', bbox_to_anchor=(1, 1))
```

```python
sns.scatterplot(data = game_data, x = 'x_coor_2', y = 'y_coor_2', hue = 'event_type', palette = event_palette)
plt.legend(title='Shot Outcome', loc='upper left', bbox_to_anchor=(1, 1))
```
